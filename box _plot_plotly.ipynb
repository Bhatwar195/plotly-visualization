## Box Plot 

A box plot is a statistical visualization used to represent the distribution of numerical data.
It shows the minimum, first quartile (Q1), median, third quartile (Q3), and maximum values.
Box plots help in understanding data spread, central tendency, and variability.
They are also useful for detecting outliers in a dataset.
Box plots are commonly used to compare distributions across different categories.

```python
import pandas as pd 
import plotly.express as px
import seaborn as sns
```

### Load Dataset 

```python
df=sns.load_dataset('tips')
```

```python
display(df.head())
```

### Example 1: Basic Box Plot

```python
fig = px.box(df, y="total_bill",x='sex', title="Total bill Vs Gender",color='sex')
fig.update_layout(
    width=1000,   # increase width
    height=600)    # increase height
fig.show()
fig.write_image("box_plot.png")
```

### Example 2: Box Plot by Category (Class-wise Marks)

```python
df2 = pd.DataFrame({
    "Class": ["A","A","A","B","B","B","C","C","C"],
    "Marks": [65, 70, 75, 55, 60, 58, 80, 85, 90]
})

fig = px.box(df2, x="Class", y="Marks",
             title="Marks Distribution by Class")
fig.update_layout(
    width=1000,   # increase width
    height=600)    # increase height
fig.show()
```

### Example 3: Box Plot with All Data Points

```python
fig = px.box(df2, x="Class", y="Marks",
             points="all",color='Class',
             title="Box Plot with Individual Data Points")
fig.update_layout(
    width=1000,   # increase width
    height=600) 
fig.show()
```

### Example 5: Gender-wise Salary Comparison

```python
df4 = pd.DataFrame({
    "Gender": ["Male","Male","Male","Female","Female","Female"],
    "Salary": [30, 35, 40, 28, 32, 38]
})

fig = px.box(df4, x="Gender", y="Salary",color='Gender',
             title="Salary Comparison by Gender")
fig.update_layout(
    width=1000,   # increase width
    height=600) 

fig.show()
```

### Example 6: Exam Score Outlier Detection

```python
df5 = pd.DataFrame({
    "Scores": [40, 45, 50, 55, 60, 65, 70, 95]
})

fig = px.box(df5, y="Scores",
             title="Box Plot Showing Outliers in Exam Scores")
fig.update_layout(
    width=1000,   # increase width
    height=600)    # increase height
fig.show()
```
