## Bar Plot 
#### A bar plot is used to compare values of different categories using rectangular bars.
#### The height of each bar represents the value of that category.

### Import Libraries

```python
import plotly.express as px
import pandas as pd 
```

### Bar Plot Example 1: Basic Bar Plot

```python
df = pd.DataFrame({
    "Day": ["Mon", "Tue", "Wed", "Thu", "Fri"],
    "Sales": [100, 150, 200, 180, 220]
})
fig = px.bar(df, x="Day", y="Sales", title="Daily Sales Bar Plot")
fig.show()
```

### Bar Plot Example 2: Bar Plot with Values on Bars

```python
fig = px.bar(
    df,
    x="Day",
    y="Sales",
    text="Sales",
    title="Daily Sales with Values"
)
fig.show()
```

### Bar Plot Example 3: Grouped Bar Plot

```python
df2 = pd.DataFrame({
    "Day": ["Mon", "Tue", "Wed", "Thu", "Fri"],
    "Product A": [120, 160, 210, 190, 230],
    "Product B": [100, 140, 180, 170, 210]
})
fig = px.bar(
    df2,
    x="Day",
    y=["Product A", "Product B"],
    title="Sales Comparison of Two Products"
)
fig.show()
```

### Bar Plot Example 4: Horizontal Bar Plot

```python
fig = px.bar(
    df,
    x="Sales",
    y="Day",
    orientation="h",
    title="Horizontal Sales Bar Plot"
)
fig.show()
```

### Bar Plot Example 5: Colored Bar Plot

```python
fig = px.bar(
    df,
    x="Day",
    y="Sales",
    color="Day",
    title="Daily Sales with Different Colors")
fig.write_image("barplot.png")   
fig.show()
```

### Bar Plot Example 6: Bar Plot with Hover Data

```python
fig = px.bar(
    df,
    x="Day",
    y="Sales",
    hover_data=["Sales"],
    title="Daily Sales with Hover Information"
)
fig.show()
```

### Bar Plot Example 7: Stacked Bar Plot

```python
df3 = pd.DataFrame({
    "Day": ["Mon", "Tue", "Wed", "Thu", "Fri"],
    "Online Sales": [60, 80, 110, 95, 130],
    "Offline Sales": [40, 70, 90, 85, 90]
})

fig = px.bar(
    df3,
    x="Day",
    y=["Online Sales", "Offline Sales"],
    title="Stacked Sales Bar Plot"
)
fig.show()
```

### Bar Plot Example 8: Bar Plot Sorted by Value

```python
df_sorted = df.sort_values(by="Sales", ascending=False)
fig = px.bar(
    df_sorted,
    x="Day",
    y="Sales",
    title="Sales Sorted in Descending Order"
)
fig.show()
```
