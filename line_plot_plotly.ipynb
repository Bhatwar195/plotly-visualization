## Line Plot 
#### A line plot is a graphical representation used to show changes or trends in data over time.
#### It connects individual data points with straight lines to make patterns and variations easy to understand.

### Import Libraries

```python
import plotly.express as px
import pandas as pd
```

### Line Plot Example 1: Basic Line Plot

```python
df1 = pd.DataFrame({
    "Day": ["Mon", "Tue", "Wed", "Thu", "Fri"],
    "Sales": [100, 150, 200, 180, 220]
})
fig = px.line(df1, x="Day", y="Sales", title="Basic Sales Trend")
fig.show()
```

### Line Plot Example 2: Line Plot with Markers

```python
fig = px.line(
    df1,
    x="Day",
    y="Sales",
    markers=True,
    title="Sales Trend with Markers"
)
fig.show()
```

### Line Plot Example 3: Multi-Line Plot

```python
df2 = pd.DataFrame({
    "Day": ["Mon", "Tue", "Wed", "Thu", "Fri"],
    "Product A": [120, 160, 210, 190, 230],
    "Product B": [100, 140, 180, 170, 210]
})
fig = px.line(
    df2,
    x="Day",
    y=["Product A", "Product B"],
    title="Comparison of Product A and Product B Sales"
)
fig.show()
```

### Line Plot Example 4: Line Plot with Hover Info

```python
fig = px.line(
    df1,
    x="Day",
    y="Sales",
    hover_data=["Sales"],
    title="Sales Trend with Hover Information"
)
fig.show()
```
